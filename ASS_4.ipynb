### Kmeans Clustering

```python
data['Genre'].replace(['Male','Female'],[1,0],inplace=True)
```

```python
data1= data[["Annual Income (k$)","Spending Score (1-100)"]]
```

```python
#data1=data.iloc[:, [3,4]].values
```

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python

kmeans = KMeans(n_clusters = 3, random_state = 0)
kmeans.fit(data1)
```

```python
kmeans.cluster_centers_
```

```python
l = kmeans.labels_
```

```python
l
```

```python
x1=data[["Annual Income (k$)"]]
```

```python
y1=data[["Spending Score (1-100)"]]
```

```python

import matplotlib.pyplot as plt

plt.scatter(x1, y1, c =l)

 
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
kmeans.inertia_
```

```python
inertias=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data1)
    inertias.append(kmeans.inertia_)
    print("Inertia Values:",kmeans.inertia_)
    
plt.plot(range(1,11), inertias, marker='o')
plt.title("Elbow Graph")
plt.xlabel("No. of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, random_state = 0)
kmeans.fit(data1)
```

```python
kmeans.cluster_centers_
```

```python
l = kmeans.labels_
```

```python
l
```

```python
plt.scatter(x1, y1, c =l)

for center in kmeans.cluster_centers_:
    center = center[:11]
    plt.scatter(center[0],center[1],marker = '^',c = 'red')
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

## Scatter Plot for Each Inertia

```python
kmeans = KMeans(n_clusters = 1, random_state = 0)
kmeans.fit(data1)
print(kmeans.cluster_centers_)
print(kmeans.labels_)
print('Inertia Value : ',kmeans.inertia_)
plt.scatter(x1, y1, c =kmeans.labels_)
for center in kmeans.cluster_centers_:
    center = center[:10]
    plt.scatter(center[0],center[1],marker = '*',c = 'red')
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```
