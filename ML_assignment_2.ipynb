```python
# Count the number of occurrences of each unique value in the 'furnishingstatus' column
furnishingstatus_counts = df["furnishingstatus"].value_counts()
furnishingstatus_labels = furnishingstatus_counts.index
plt.pie(furnishingstatus_counts, labels=furnishingstatus_labels,  autopct='%.1f%%', startangle=140)
plt.title("Distribution of furnishingstatus")
# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis("equal")
plt.show()
```

```python
# Count the number of occurrences of each unique value in the 'bathroom' column
bathroom_counts = df["bedrooms"].value_counts()
bathroom_labels = bathroom_counts.index
plt.pie(bathroom_counts, labels=bathroom_labels,  autopct='%.1f%%', startangle=140)
plt.title("Distribution of bathrooms")

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

plt.show()
```

```python
df["furnishingstatus"]= df["furnishingstatus"].map({"unfurnished":0, "semi-furnished":1, "furnished":2})
df["mainroad"]= df["mainroad"].map({"no":0, "yes":1})
df["guestroom"]= df["guestroom"].map({"no":0, "yes":1})
df["basement"]= df["basement"].map({"no":0, "yes":1})
df["hotwaterheating"]= df["hotwaterheating"].map({"no":0, "yes":1})
df["airconditioning"]= df["airconditioning"].map({"no":0, "yes":1})
df["prefarea"]= df["prefarea"].map({"no":0, "yes":1})
```

```python
x=df[["area"]]
y=df[["price"]]
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75)

print(x_train.shape)
print(x_train)
```

```python
print(y_train.shape)
print(y_train)
```

```python
print(x_test.shape)
print(x_test)
```

```python
print(y_test.shape)
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
predicted = LR.predict(x_test)
print(predicted)
```

```python
LR.predict([[10000]])
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,predicted)
```

```python
from math import sqrt
rmse = sqrt(mean_squared_error(y_test, predicted))
print(rmse)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test,predicted)
print(mae)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,predicted)
```
