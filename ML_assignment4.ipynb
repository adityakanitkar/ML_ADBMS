```python
df["Genre"]=df["Genre"].map({"Male":0,"Female":1,"I":2})
```

```python
x=df.iloc[:, 3:].values
x
```

```python
plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"])
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0) 
kmeans.fit_predict(x)  
```

```python
cent = kmeans.cluster_centers_
cent
```

```python
kmeans.inertia_
```

```python
labels = kmeans.labels_
labels
```

```python
plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"], c=labels)
plt.scatter(cent[:, 0],cent[:, 1], marker='*',color='r')
```

```python
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit_predict(x)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
km = KMeans(n_clusters=5,random_state=1)
km.fit_predict(x)
```

```python
labels = km.labels_
labels
```

```python
km.inertia_
```

```python
cent1 = km.cluster_centers_
cent1
```

```python
plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"], c=labels)
plt.scatter(cent1[:, 0] , cent1[:, 1], s=100, marker= '*' ,color='r')
```

```python
df[labels==0]
```

```python
fit=km.fit_predict(x)

plt.scatter(x[fit == 0,0] , x[fit == 0,1] , s=100 , c='red',label='cluster 1')
plt.scatter(x[fit ==1,0], x[fit==1,1],s=100, c='green', label='cluster 2')
plt.scatter(x[fit ==2,0], x[fit==2,1],s=100, c='pink', label='cluster 3')
plt.scatter(x[fit ==3,0], x[fit==3,1],s=100, c='blue', label='cluster 4')
plt.scatter(x[fit ==4,0], x[fit==4,1],s=100, c='orange', label='cluster 5')

plt.scatter(cent1[:, 0] , cent1[:, 1], s=200, marker= '*' ,color='k')

plt.legend()
plt.show()
```

```python
km.predict([[46, 78]])
```
