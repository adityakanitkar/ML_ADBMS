```python
x=df.drop("Admitted",axis=1)
y=df["Admitted"]
```

```python
import sklearn.model_selection
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.75)
```

```python
print(x_train.shape)
print(x_train)
```

```python
print(y_train.shape)
print(y_train)
```

```python
print(x_test.shape)
x_test
```

```python
print(y_test.shape)
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0,max_depth=3)
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
y_pred=y_pred.astype('int64')
```

```python
result=pd.DataFrame({"actual":y_test,"predicted":y_pred})
result
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,accuracy_score,classification_report
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(clf)
```

```python
plt.figure(figsize=(10,10))
features = ["GRE Score",	"TOEFL Score",	"University Rating",	"SOP",	"LOR",	"CGPA",	"Research"]
plot_tree(clf,fontsize=8,filled=True,feature_names=features,class_names=['NotAdmitted','Admitted'])
```

```python
dtr = DecisionTreeClassifier(criterion="entropy",random_state=0,max_depth=3)
dtr.fit(x_train,y_train)
y_pred = dtr.predict(x_test)
y_pred=y_pred.astype('int64')
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
plt.figure(figsize=(10,10))
plot_tree(dtr,fontsize=8,filled=True,feature_names=features,class_names=['NotAdmitted','Admitted'])
```
